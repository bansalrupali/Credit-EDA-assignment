#### Merge both data sheets

```python
Merged_Data = pd.merge(df, df1, how='inner', on=['SK_ID_CURR'])
Merged_Data.head()
```

```python
Merged_Data.info()
```

```python
plt.figure(figsize=(18, 8))
sns.barplot(x=Merged_Data['NAME_CONTRACT_STATUS'],y=Merged_Data['TARGET'])
plt.xticks(rotation = 90)
plt.show()
```

```python
Merged = Merged_Data.pivot_table(values='SK_ID_CURR', 
                      index='NAME_EDUCATION_TYPE',
                      columns='NAME_CONTRACT_STATUS',
                      aggfunc='count')

Merged=Merged.div(Merged.sum(axis=1),axis='rows')*100
Merged.plot(kind='bar',figsize=(15,5))
```

```python
Merged = Merged_Data.pivot_table(values='SK_ID_CURR', 
                      index='AMT_INCOME_TOTAL_RANGE',
                      columns='NAME_CONTRACT_STATUS',
                      aggfunc='count')

Merged=Merged.div(Merged.sum(axis=1),axis='rows')*100
Merged.plot(kind='bar',figsize=(15,5))
```
